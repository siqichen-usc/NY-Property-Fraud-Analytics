```python
import pandas as pd
import numpy as np
import scipy.stats as sps 
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn as skl
from sklearn import preprocessing 
%matplotlib inline
```

```python
%%time
data = pd.read_csv('NYP_45_var_revised.csv')
```

```python
normalized_data = (data - data.mean()) / data.std()
```

```python
normalized_data.head()
```

```python
normalized_data.drop('Unnamed: 0',axis=1, inplace=True)
normalized_data.head()
```

```python
normalized_data.describe()
```

```python
from sklearn.decomposition import PCA
```

```python
features = ['FULLVAL_r1', 'FULLVAL_r2', 'FULLVAL_r3', 'AVLAND_r4', 'AVLAND_r5',
       'AVLAND_r6', 'AVTOT_r7', 'AVTOT_r8', 'AVTOT_r9', 'r1_byzip5',
       'r1_byzip3', 'r1_bytaxclass', 'r1_b', 'r2_byzip5', 'r2_byzip3',
       'r2_bytaxclass', 'r2_b', 'r3_byzip5', 'r3_byzip3', 'r3_bytaxclass',
       'r3_b', 'r4_byzip5', 'r4_byzip3', 'r4_bytaxclass', 'r4_b', 'r5_byzip5',
       'r5_byzip3', 'r5_bytaxclass', 'r5_b', 'r6_byzip5', 'r6_byzip3',
       'r6_bytaxclass', 'r6_b', 'r7_byzip5', 'r7_byzip3', 'r7_bytaxclass',
       'r7_b', 'r8_byzip5', 'r8_byzip3', 'r8_bytaxclass', 'r8_b', 'r9_byzip5',
       'r9_byzip3', 'r9_bytaxclass', 'r9_b']
x = data.loc[:,features].values
```

```python
pca = PCA(n_components = 7)
principalComponents = pca.fit_transform(normalized_data)
DF = pd.DataFrame(data = principalComponents)
```

```python
DF.shape
```

```python
DF.head()
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```

```python
finalDF = (DF-DF.mean())/DF.std()
finalDF.head()
```

```python
newDF = abs(finalDF)
newDF.head()
```

```python
newDF['Score1'] = newDF.transpose().sum()
newDF.head()
```

```python
newDF['Score2'] = pow(finalDF,2).transpose().sum()
```

```python
newDF['Score2'] = pow(newDF['Score2'],0.5)
newDF.head()
```

```python
newDF['RECORD'] = newDF.index + 1 
```

```python
newDF.sort_values('Score2',ascending = False).head()
```

```python
newDF.reset_index()
```

```python
newDF0=newDF.reset_index()
newDF0['index']=newDF0['index']+1
```

```python
zscore = newDF0[['RECORD','Score1','Score2']]
```
